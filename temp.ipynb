```python
from flask import Flask, render_template, request,jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
import logging
```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + "iphone16"
flipkart_url
```

```python
uClient = uReq(flipkart_url)
uClient
```

```python
flipkartPage = uClient.read()
flipkartPage
```

```python
flipkart_html = bs(flipkartPage, "html.parser")
```

```python
flipkart_html
```

```python
bigboxes = flipkart_html.find_all("div", {"class": "cPHDOP col-12-12"})
```

```python
bigboxes
```

```python
del bigboxes[0:2]
box = bigboxes[0]
```

```python
box
```

```python
productLink = "https://www.flipkart.com" + box.div.div.div.a['href']
```

```python
productLink
```

```python
prodRes = requests.get(productLink)
```

```python
prodRes
```

```python
prodRes.encoding='utf-8'
prod_html = bs(prodRes.text, "html.parser")
print(prod_html)
```

```python
commentboxes = prod_html.find_all('div', {'class': "RcXBOT"})
```

```python
commentboxes
```

```python
name = commentboxes[0].div.div.find_all('p', {'class': '_2NsDsF AwS1CA'})[0].text
name
```

```python
reviews = []  # list to store all review dicts

for commentbox in commentboxes:
    if commentbox.div and commentbox.div.div:
        name_tag = commentbox.div.div.find_all('p', {'class': '_2NsDsF AwS1CA'})
        if name_tag:
            name = name_tag[0].text.strip()
            mydict = {
                "Product": "iphone16",
                "Name": name
            }
            reviews.append(mydict)
```
